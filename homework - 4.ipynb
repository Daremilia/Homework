```python
def get_coef(y, x, weight=1):
    if type(weight).__name__ == "int":
        return np.linalg.inv((x.T @ x)) @ (x.T @ y)
    else:
        weight = np.linalg.inv(weight)
        return np.linalg.inv((x.T @ weight @ x)) @ x.T @ weight @ y
    
def get_σ2(y, x, weight=1):
    β_head = get_coef(y, x, weight)
    ϵ_head = y - x @ β_head
    return np.sum(ϵ_head ** 2) / N

def get_R2(y, x, weight=1):
    β_head = get_coef(y, x, weight)
    ϵ_head = y - x @ β_head
    R2 = 1 - np.sum(ϵ_head ** 2) / np.sum((y - y.mean()) ** 2)
    return R2

def get_I(y_vec, x_mat, β, σ2):
    I = np.zeros((4,4))
    for ii in range(N):
        xi = x_mat[ii:ii+1]
        yi = y_vec[ii]
        temp1 = xi.T@(yi-xi@β)/σ2
        temp2 = -1/(2*σ2)+(yi-xi@β)**2/(2*(σ2**2))
        gi = np.vstack((temp1, temp2))
        I += gi@gi.T
    return I

def get_g(y_vec, x_mat, β, σ2):
    g = np.zeros((4,1))
    for ii in range(N):
        xi = x_mat[ii:ii+1]
        yi = y_vec[ii]
        temp1 = xi.T@(yi-xi@β)/σ2
        temp2 = -1/(2*σ2)+((yi-xi@β)**2)/(2*(σ2**2))
        gi = np.vstack((temp1, temp2))
        g += gi
    return g
```
